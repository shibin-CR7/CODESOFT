```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df = pd.read_csv('/content/spam.csv', encoding='iso-8859-1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
```

```python
df.rename(columns = {'v1':'Category', 'v2':'Message'}, inplace = True)
```

```python
spam = df.groupby('Category')['Message'].count()
spam
```

```python
label_encoder = preprocessing.LabelEncoder()
df['spam'] = label_encoder.fit_transform(df['Category'])
df.head()
```

```python
x = df['Message']
y = df['spam']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

```python
vect = CountVectorizer()
x_train_count = vect.fit_transform(x_train.values)
x_train_count.toarray()[:3]
```

```python
model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
emails = ['Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!']
emails_count = vect.transform(emails)
model.predict(emails_count)
```

```python
count = vect.transform(x_test)
model.predict(count)
```

```python
model.score(count, y_test)
```

```python
pred = model.predict(count)
conf_m = confusion_matrix(y_test,pred)
conf_m
```

```python
cslfr=classification_report(y_test,pred)
print(cslfr)
```
